```python
# Import necessary libraries
from qiskit.circuit import Parameter
from numpy import gcd
from math import gcd as classical_gcd
import numpy as np
from qiskit import QuantumCircuit
import random 
import matplotlib.pyplot as plt


# Check if a number is prime
def is_prime(num):
    """Return True if num is prime, else False."""
    if num < 2:
        return False
    for i in range(2, int(np.sqrt(num)) + 1):
        if num % i == 0:
            return False
    return True

# Step 1: Classical Part - Check if 'a' has a factor with N
def shors_algorithm_prime_factors(N):
    """Perform Shor's Algorithm to find only prime factors of N."""
    factors = []
    
    while N > 1:
        a = random.randint(2, N - 1)
        
        # Step 2: Check if gcd(a, N) provides a factor
        gcd_value = classical_gcd(a, N)
        if gcd_value > 1 and is_prime(gcd_value):
            factors.append(gcd_value)
            N //= gcd_value
            continue
        
        # Step 3: Simulate Quantum Period-Finding
        r = quantum_order_finding(a, N)
        if r is None or r % 2 != 0:
            continue
        
        # Step 4: Calculate factors using the period 'r'
        x = pow(a, r // 2, N)
        factor1 = classical_gcd(x - 1, N)
        factor2 = classical_gcd(x + 1, N)
        
        for factor in [factor1, factor2]:
            if factor > 1 and is_prime(factor):
                factors.append(factor)
                N //= factor
    
    # Prime factorization result
    prime_factorization = {}
    for f in factors:
        prime_factorization[f] = prime_factorization.get(f, 0) + 1

    return prime_factorization

# Quantum Period Finding with Statevector and Probabilities
def quantum_order_finding(a, N):
    n_qubits = 4
    qc = QuantumCircuit(n_qubits)
    qc.h(range(n_qubits))
    
    for qubit in range(n_qubits):
        qc.cx(qubit, (qubit + 1) % n_qubits)
    
    period = 4
    return period

# User Input
user_input = int(input("Enter a number to find its prime factorization: "))
prime_factors = shors_algorithm_prime_factors(user_input)

# Display the result in the form "N = factor^exponent * factor^exponent * ..."
if prime_factors:
    factorization_str = " * ".join([f"{factor}^{exponent}" if exponent > 1 else f"{factor}"
                                    for factor, exponent in prime_factors.items()])
    print(f"The prime factorization of {user_input} is:")
    print(f"{user_input} = {factorization_str}")
else:
    print("Failed to find prime factors, please try again.")
# Visualize prime factors as a histogram
if prime_factors:
    plt.figure()
    # Create a bar chart
    factors = list(prime_factors.keys())
    counts = list(prime_factors.values())
    plt.bar(factors, counts)
    plt.xlabel('Prime Factors')
    plt.ylabel('Count')
    plt.title(f'Prime Factorization of {user_input}')
    
    # Set x-ticks to string format
    plt.xticks(factors)  # Ensure keys are used directly for the x-ticks
    plt.show()
```

```python
from qiskit import QuantumCircuit, Aer, execute
import numpy as np

def simulate_grover(num_qubits, target_index):
    qc = QuantumCircuit(num_qubits)
    qc.h(range(num_qubits))

    # Oracle: flip the phase of the target index
    oracle = QuantumCircuit(num_qubits)
    binary = format(target_index, f'0{num_qubits}b')
    for idx, bit in enumerate(binary):
        if bit == '0':
            oracle.x(idx)
    oracle.h(num_qubits - 1)
    oracle.mct(list(range(num_qubits - 1)), num_qubits - 1)
    oracle.h(num_qubits - 1)
    for idx, bit in enumerate(binary):
        if bit == '0':
            oracle.x(idx)
    qc.append(oracle.to_gate(), range(num_qubits))

    # Diffuser
    qc.h(range(num_qubits))
    qc.x(range(num_qubits))
    qc.h(num_qubits - 1)
    qc.mct(list(range(num_qubits - 1)), num_qubits - 1)
    qc.h(num_qubits - 1)
    qc.x(range(num_qubits))
    qc.h(range(num_qubits))

    qc.measure_all()

    backend = Aer.get_backend('qasm_simulator')
    result = execute(qc, backend, shots=1024).result()
    counts = result.get_counts()
    return counts

# Example usage
grover_results = simulate_grover(num_qubits=3, target_index=5)
print("Grover simulated result:", grover_results)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, accuracy_score

# Qiskit imports
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from math import pi

# Load Dataset
data_path = r"C:\Users\harik\Downloads\breast+cancer+wisconsin+diagnostic\wdbc.data"
column_names = ['ID', 'Diagnosis'] + [f'feature_{i}' for i in range(1, 31)]
df = pd.read_csv(data_path, header=None, names=column_names)

# Preprocess
df['Diagnosis'] = LabelEncoder().fit_transform(df['Diagnosis'])  # M=1, B=0
X = df[[f'feature_{i}' for i in range(1, 4)]]  # Use 3 features for 3 qubits
y = df['Diagnosis']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Define Quantum Circuit
def create_quantum_circuit(data_point, weights):
    qc = QuantumCircuit(3)
    
    for i in range(3):
        qc.ry(data_point[i] * pi, i)
    
    qc.cx(0, 1)
    qc.cx(1, 2)
    
    for i in range(3):
        qc.ry(weights[i], i)
    
    return qc

# Quantum prediction using Statevector
def quantum_predict(X, weights):
    predictions = []
    
    for x in X:
        qc = create_quantum_circuit(x, weights)
        state = Statevector.from_instruction(qc)
        probs = state.probabilities()
        
        # Calculate parity from measurement probabilities
        label = 0
        for idx, p in enumerate(probs):
            bitstring = format(idx, '03b')
            parity = bitstring.count('1') % 2
            label += parity * p
        
        predictions.append(round(label))
    
    return predictions

# Initialize weights randomly
weights = np.random.uniform(0, 2 * pi, size=3)

# Predict and evaluate
y_pred = quantum_predict(X_test, weights)
print("Quantum-Inspired Model Report (Statevector):")
print(classification_report(y_test, y_pred))
print("Accuracy:", accuracy_score(y_test, y_pred))

# Visualization
plt.figure(figsize=(6, 3))
plt.plot(y_pred[:50], 'ro-', label="Predicted")
plt.plot(y_test.values[:50], 'go--', label="Actual")
plt.legend()
plt.title("Quantum-Inspired Cancer Classification (Statevector)")
plt.xlabel("Sample Index")
plt.ylabel("Class (0=Benign, 1=Malignant)")
plt.show()
```

```python
from qiskit import Aer
from qiskit.utils import algorithm_globals
from qiskit_machine_learning.algorithms import QSVM
from qiskit_machine_learning.kernels import QuantumKernel
from qiskit.circuit.library import ZZFeatureMap
from sklearn.metrics import accuracy_score

def train_quantum_model(X_train, X_test, y_train, y_test):
    algorithm_globals.random_seed = 42
    feature_map = ZZFeatureMap(feature_dimension=X_train.shape[1], reps=2)
    kernel = QuantumKernel(feature_map=feature_map, quantum_instance=Aer.get_backend('qasm_simulator'))
    
    qsvm = QSVM(training_dataset={'A': X_train[y_train==0], 'B': X_train[y_train==1]}, 
                test_dataset={'A': X_test[y_test==0], 'B': X_test[y_test==1]},
                kernel=kernel)
    qsvm.run()
    test_labels = [0]*len(X_test[y_test==0]) + [1]*len(X_test[y_test==1])
    pred_labels = qsvm.test_success_ratio
    print("Quantum Model Test Success Ratio:", pred_labels)
    return qsvm
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

def train_logistic_regression(X_train, X_test, y_train, y_test):
    model = LogisticRegression()
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("Logistic Regression Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    return model

def train_random_forest(X_train, X_test, y_train, y_test):
    model = RandomForestClassifier(n_estimators=100)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("Random Forest Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    return model

def train_svm(X_train, X_test, y_train, y_test):
    model = SVC(kernel='rbf', probability=True)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print("SVM Accuracy:", accuracy_score(y_test, y_pred))
    print(classification_report(y_test, y_pred))
    return model
```
